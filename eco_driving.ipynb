# 데이터 전처리

```python
# s20부터 s150까지의 열 이름 리스트 생성
columns = ['s' + str(i) for i in range(20, 151, 10)]
columns.append('idle_time')

# 해당 열의 값들을 리스트로 변환하고 각 행의 표준편차 계산
df_1['use_fuel'] = df_1['s_fuel']-df_1['e_fuel']
df_1['distance'] = (df_1['e_odo']-df_1['s_odo'])/1000
df_1['idle_time'] = df_1['idle_time'].apply(lambda x: 0 if x < 0 else x)
df_1['total_s'] = df_1.loc[:, 's20':'s150'].sum(axis=1)
df_1['total_s'] = df_1['total_s']+df_1['idle_time']

df_1['eco_s_m'] = df_1.loc[:, 's60':'s70'].sum(axis=1)

df_1['eco_s_l'] = df_1.loc[:, ['s60','s70','s90']].sum(axis=1)

```

```python
# s_date 열을 날짜 형식으로 변환
df_1['s_date'] = pd.to_datetime(df_1['s_date'])
df_1['e_date'] = pd.to_datetime(df_1['e_date'])

# 날짜를 YYYY-MM-DD 형식으로 포맷
df_1['year_month_date_s'] = df_1['s_date'].dt.strftime('%Y-%m-%d')
df_1['year_month_date_e'] = df_1['e_date'].dt.strftime('%Y-%m-%d')

df_1['year_month_date_s']
```

# 예약 아이디별로 Feature 생성

```python
df_grouped = df_1.groupby(['reservation_id2', 'cmp_svc_id']).agg(
    distance_sum=('distance', 'sum'),
    idle_time_sum=('idle_time', 'sum'),
    total_s_sum=('total_s', 'sum'),
    eco_s_m_sum=('eco_s_m', 'sum'),
    eco_s_l_sum=('eco_s_l', 'sum'),
    use_fuel_sum=('use_fuel', 'sum'),
    avg_speed=('avg_spd', 'mean'),
    max_speed=('max_spd', 'max')
).reset_index()
```

```python
df_grouped.head()
```

```python
df_grouped['variance_speed'] = (df_grouped['max_speed']-df_grouped['avg_speed'])/df_grouped['max_speed']
df_grouped['fuel_eff'] = df_grouped['distance_sum']/df_grouped['use_fuel_sum']
df_grouped['normalized_eco_drive_s_m'] = df_grouped['eco_s_m_sum']/df_grouped['total_s_sum']
df_grouped['normalized_eco_drive_s_l'] = df_grouped['eco_s_l_sum']/df_grouped['total_s_sum']
df_grouped['normalized_idle'] = (df_grouped['idle_time_sum']/df_grouped['total_s_sum'])
```

```python
df_grouped.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df_grouped.head()
```

# 이상치 제거

```python

df_filtered = df_grouped[(df_grouped['use_fuel_sum'] > 0) & (df_grouped['variance_speed']>0) & (df_grouped['distance_sum']>0) & (df_grouped['variance_speed']>0) & (df_grouped['normalized_idle']!=1) & (df_grouped['total_s_sum']>0)&
                         (df_grouped['normalized_eco_drive_s_m']<1)&(df_grouped['normalized_eco_drive_s_l']<1)]
```

# 에코스코어 생성

```python
df_filtered_1['variance_speed_score'] = 1 - ((df_filtered_1['variance_speed']))
df_filtered_1['eco_drive_s_m_score'] = df_filtered_1['normalized_eco_drive_s_m']
df_filtered_1['eco_drive_s_l_score'] = df_filtered_1['normalized_eco_drive_s_l']
df_filtered_1['idle_score'] = 1 -  (df_filtered_1['normalized_idle'])
```

```python
# Weights for each feature
weights_long = [0.2, 0.60, 0.2]
weights_medium = [0.20, 0.60, 0.2]
weights_short = [0.33, 0.33, 0.33]

# Calculate scores based on conditions
score_long = (df_filtered_1['variance_speed_score'] * weights_long[0] +
              df_filtered_1['eco_drive_s_l_score'] * weights_long[1] +
              df_filtered_1['idle_score'] * weights_long[2])

score_medium = (df_filtered_1['variance_speed_score'] * weights_medium[0] +
                df_filtered_1['eco_drive_s_m_score'] * weights_medium[1] +
                df_filtered_1['idle_score'] * weights_medium[2])

score_short = (df_filtered_1['variance_speed_score'] * weights_short[0] +
               df_filtered_1['eco_drive_s_m_score'] * weights_medium[1] +
               df_filtered_1['idle_score'] * weights_short[2])

# Assign scores based on conditions
df_filtered_1['eco_driving_score'] = np.where(df_filtered_1['distance_sum'] > 100, score_long,
                                         np.where(df_filtered_1['distance_sum'] > 50, score_medium, score_short))
```
